# Question No. 21(a)

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars=pd.read_csv('cars.csv')
```

```python
cars.head()
```

```python
cars['MPG'].mean()
```

```python
cars['MPG'].median()
```

```python
cars['MPG'].mode()
```

```python
cars['MPG'].hist()
```

```python
sns.distplot(cars['MPG'])
plt.grid(True)
plt.show()
```

```python
cars['MPG'].skew()
```

```python
cars['MPG'].kurt()
```

```python
#From above plot and values we can say that data is fairly symmetrical, i.e fairly normally distributed.
```

# Question No. 21(b)

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('wc-at.csv')
```

```python
df.head()
```

```python
df.mean()
```

```python
df.median()
```

```python
df.mode()
```

```python
sns.distplot(df['Waist'])
plt.show()
```

```python
sns.distplot(df['AT'])
plt.show()
```

```python
sns.boxplot(df['AT'])
plt.show()

# mean> median, right whisker is larger than left whisker, data is positively skewed.
```

```python
sns.boxplot(df['Waist'])
plt.show()

## mean> median, both the whisker are of same length, median is slightly shifted towards left. Data is fairly symmetrically distributed.
```
